```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
time = 0

def insertionSort(n_lista):
    global time
    for index in range(1, len(n_lista)):
        time += 1
        current = n_lista[index]
        position = index
        while position > 0 and n_lista[position - 1] > current:
            time += 1
            n_lista[position] = n_lista[position - 1]
            position = position - 1
        n_lista[position] = current
    return n_lista

def mergeSort(n_lista):
    if len(n_lista) <= 1:
        return n_lista
    
    mid = len(n_lista) // 2
    left_half = n_lista[:mid]
    right_half = n_lista[mid:]

    left_half = mergeSort(left_half)
    right_half = mergeSort(right_half)

    return merge(left_half, right_half)

def merge(left_half, right_half):
    merged = []
    left_index = 0
    right_index = 0

    while left_index < len(left_half) and right_index < len(right_half):
        if left_half[left_index] <= right_half[right_index]:
            merged.append(left_half[left_index])
            left_index += 1
        else:
            merged.append(right_half[right_index])
            right_index += 1

    while left_index < len(left_half):
        merged.append(left_half[left_index])
        left_index += 1

    while right_index < len(right_half):
        merged.append(right_half[right_index])
        right_index += 1

    return merged

def quickSort(n_lista):
    if len(n_lista) <= 1:
        return n_lista
    
    pivot = n_lista[len(n_lista) // 2]
    left = [x for x in n_lista if x < pivot]
    middle = [x for x in n_lista if x == pivot]
    right = [x for x in n_lista if x > pivot]

    return quickSort(left) + middle + quickSort(right)

TAM = 101
eje_x = list(range(1, TAM, 1))
eje_y_insertion = []
eje_y_merge = []
eje_y_quick = []

for num in eje_x:
    lista_variable = random.sample(range(0, 1000), num)
    print("Unsorted List:", lista_variable)

    time = 0
    insertion_sorted = insertionSort(lista_variable.copy())
    eje_y_insertion.append(time)
    print("Insertion Sorted List:", insertion_sorted)

    merge_sorted = mergeSort(lista_variable.copy())
    eje_y_merge.append(len(merge_sorted))
    print("Merge Sorted List:", merge_sorted)

    quick_sorted = quickSort(lista_variable.copy())
    eje_y_quick.append(len(quick_sorted))
    print("Quick Sorted List:", quick_sorted)

    print()

fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(eje_x, eje_y_insertion, marker="o", color="b", linestyle='None')
ax.plot(eje_x, eje_y_quick, marker="o", color="r", linestyle='None')
ax.plot(eje_x, eje_y_merge, marker="x", color="g", linestyle='None')


ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True)
ax.legend(["Insertion Sort", "Merge Sort", "Quick Sort"])

plt.title("Sorting Algorithm Performance")
plt.show()

```
